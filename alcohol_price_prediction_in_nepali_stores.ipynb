```python
predictors = alcohol_data[['Volume','Alcohol','en_category']]
target = alcohol_data[['Price']]
```

```python
import numpy as np
```

```python
predictors_array = np.asarray(predictors)
```

```python
predictors_array
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors_array,target_array,test_size=0.2,random_state=3)
```

```python
predictions = lr.predict(x_test)
```
